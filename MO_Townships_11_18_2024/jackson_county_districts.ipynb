# Jackson County Missouri Townships

We received a request asking for shapes for various districts for 5 different cities / townships in Jackson County MO.

Independence: https://www.independencemo.gov/government/city-departments/city-councilcity-clerk/about-city-council
Lee's Summit: https://www.arcgis.com/home/webmap/viewer.html?webmap=79388b80bbd04911ae59cb57ab513061
Raytown: https://www.raytown.mo.us/index.asp?SEC=5B3A5C73-F34B-411D-B35D-12FA011FAFEB
Grandview: https://www.grandview.org/home/showpublisheddocument/13743/638651130004300000
Grain Valley: https://www.cityofgrainvalley.org/files/administration/ward_map_20220729_95799.pdf

```python
mo_geos = gp.read_file("/Users/peterhorton/Documents/RDH/Support/Processing-Requests/MO_Townships_11_18_2024/raw-from-source/selected_mo_places_splits/selected_mo_places_splits.shp")
```

```python
mo_geos
```

```python
mo_geos = mo_geos[["NAME","NAMELSAD", "geometry"]]
```

```python
mo_geos
```

```python
independence_mo = gp.read_file("./raw-from-source/independence_mo_districts.geojson")
```

```python
lee_districts = gp.read_file("./raw-from-source/lee_summit_council_districts.geojson")
```

```python
independence_mo["NAMELSAD"] = "Independence"
```

```python
lee_districts["NAME"] = lee_districts["CMEMBER1"].map({"Beto Lopez":3,"Mia Prier":1,"Fred DeMoro":4,"John Lovell":2})
```

```python
lee_districts["NAMELSAD"] = "Lee"
```

```python
lee_districts = lee_districts[["NAME", "NAMELSAD", "geometry"]]
independence_mo = independence_mo[["NAME", "NAMELSAD", "geometry"]]
```

```python
lee_districts = lee_districts.to_crs(mo_geos.crs)
independence_mo = independence_mo.to_crs(mo_geos.crs)
```

```python
independence_mo = independence_mo[independence_mo["NAME"].str.contains("District")]
```

```python
comb = gp.GeoDataFrame(pd.concat([mo_geos, lee_districts, independence_mo]))
```

```python
comb["NAMELSAD"] = comb["NAMELSAD"].str.replace(" city", "")
```

```python
comb["Type"] = ""
```

```python
comb.loc[comb["NAMELSAD"]=="Independence","Type"] = "City Council"
comb.loc[comb["NAMELSAD"]=="Lee","Type"] = "City Council"
comb.loc[comb["NAMELSAD"]=="Raytown","Type"] = "Ward"
comb.loc[comb["NAMELSAD"]=="Grandview","Type"] = "Ward"
comb.loc[comb["NAMELSAD"]=="Grain Valley","Type"] = "Ward"
```

```python
comb["NAME"] = comb["NAME"].apply(lambda x:str(x).split(" ")[-1])
```

```python
comb.rename(columns = {"NAME":"District","NAMELSAD":"NAME"}, inplace = True)
```

```python
comb = comb[["NAME","Type","District","geometry"]]
```

```python
comb.sort_values(["NAME", "District"], inplace = True)
```

```python
comb.reset_index(inplace = True, drop = True)
```

```python
for name in list(comb["NAME"].unique()):
    comb[comb["NAME"]==name].plot(column = "District", legend = True)
```
