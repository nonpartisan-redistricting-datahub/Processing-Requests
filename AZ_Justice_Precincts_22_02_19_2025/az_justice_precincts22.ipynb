```python
import geopandas as gp
import pandas as pd
import maup
```

# Arizona Justice Precincts with Native Population and Partisanship Estimates

```python
# Load Justice Precincts, Create Unique ID field and Change Indice for Maup
az_justice_precs = gp.read_file("./raw-from-source/az_justice_peace_precincts.geojson")
az_justice_precs["DistNum"] = az_justice_precs["DistNum"].fillna(1)
az_justice_precs["DistNum"] = az_justice_precs["DistNum"].astype(float).astype(int).astype(str)
az_justice_precs["UNIQUE_ID"] = az_justice_precs["BdName"] + "-" + az_justice_precs["DistNum"]
az_justice_precs.index = az_justice_precs["UNIQUE_ID"]
```

```python
# Load 2020 Blocks with Demographic Data
BLOCK_PATH = "/Users/peterhorton/Downloads/az_pl2020_b/az_pl2020_b.shp"
az_blocks = gp.read_file(BLOCK_PATH)
```

```python
# Assign blocks to justice precincts
az_blocks = az_blocks.to_crs(az_justice_precs.crs)
az_blocks["justice_prec"] = maup.assign(az_blocks, az_justice_precs)
```

```python
# Load blocks with 2020 election data, merge data
NATIONAL_BLOCK_PATH = "/Users/peterhorton/Downloads/national_block_2020_pres_results/national_block_2020_pres_results.csv"
national_blocks = pd.read_csv(NATIONAL_BLOCK_PATH)
national_blocks["GEOID20"] = national_blocks["GEOID20"].astype(str).str.zfill(15)

az_blocks = pd.merge(az_blocks,national_blocks,how="left",on="GEOID20",indicator=True)
```

```python
# Aggregate blocks to the justice-precinct level, add election data of two-party presidential vote share
justice_precs_data = az_blocks.dissolve(by = "justice_prec", aggfunc = "sum", as_index = False)
justice_precs_data["REP_EST20"] = round(justice_precs_data["G20PRERTRU"] / (justice_precs_data["G20PRERTRU"]+justice_precs_data["G20PREDBID"]),3)
justice_precs_data["DEM_EST20"] = round(justice_precs_data["G20PREDBID"] / (justice_precs_data["G20PRERTRU"]+justice_precs_data["G20PREDBID"]),3)
```

```python
# Add a county field to the justice precincts
az_cnty = gp.read_file("/Users/peterhorton/Documents/RDH/raw_data/census/2020_TIGER_CNTY/az_cnty_2020_bound/az_cnty_2020_bound.shp")
az_cnty.index = az_cnty["NAME20"]
az_cnty = az_cnty.to_crs(justice_precs_data.crs)
justice_precs_data["County"] = maup.assign(justice_precs_data, az_cnty)
```

```python
# Load in csv of field names to get American Indian Fields
fields = pd.read_csv("/Users/peterhorton/Documents/RDH/Fields_and_Descriptions.csv")
amer_ind = set(fields[(fields["Table"]=="P1")&(fields["Table number and contents"].str.contains("American Indian"))]["Data dictionary reference name"])

# Add any part native total population and percent fields
justice_precs_data["TOT_NTV20"] = justice_precs_data[amer_ind].sum(axis = 1)
justice_precs_data["NTV_PCT20"] = round(justice_precs_data["TOT_NTV20"] / justice_precs_data["P0010001"],3) 
```

```python
justice_precs_data.rename(columns = {"TOT_POP":"TOT_POP20","justice_prec":"JUS_PREC22",'County':"COUNTY"}, inplace = True)
justice_precs_data = justice_precs_data[["JUS_PREC22","COUNTY","TOT_POP20","TOT_NTV20","NTV_PCT20","REP_EST20","DEM_EST20","geometry"]]

import os
if not os.path.exists("./az_justice_precs_22/"):
    os.mkdir("./az_justice_precs_22/")

if not os.path.exists("./az_justice_precs_22/az_justice_precs_22"):
    os.mkdir("./az_justice_precs_22/az_justice_precs_22/")
    
justice_precs_data.to_file("./az_justice_precs_22/az_justice_precs_22/az_justice_precs_22.shp")
justice_precs_data.drop(["geometry"], axis =1).to_csv("./az_justice_precs_22/az_justice_precs_22.csv", index = False)
```
